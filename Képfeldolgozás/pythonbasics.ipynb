###__1.1. Types__

```python
# This is a comment.

# value assignment
x = 1
# printing
print(x)
# checking types
print(type(x))
```

###__1.2. Data structures__

```python
# List and tuple
a = [1, 2, 3]
b = (1, 2, 3)

print()  # Blank line
print(a, ":", len(a))
print(b, ":", len(b))
```

```python
# Indexing is the same.
print(a[0])
print(b[0])
```

```python
# Multidimensional arrays can be created with the embedding of brackets.
a = [[1, 2, 3, 4], [5, 6, 7, 8]]
print(a)
# Indexing
print(a[0][1])
```

```python
# Elements with different types are also possible in the same data structure
a = ['a', 2, 3.0, True]
b = ('b', 4, 0.1, False)

print(a)
print(b)
```

```python
# Dictionary: Storing key-value pairs
c = {1 : 2, 'a' : 'b', True : False}

print(c)

# indexing with keys
print(c[1])
print(c['a'])
print(c[True])
print()
```

```python
# What is the problem? Why do we get the same using 1 and True as indices?
print("Are True and 1 the same? Answer", True == 1)
# Based on that:
print("Are False and 0 the same? Answer:", False == 0)
```

```python
# Solution: Don't use True and 1, nor False and 0 values as keys in the same dict.
c = {
    3 : 2,
    'a' : 'b',
    True : False}

print(c)
```

###__1.3. Conditional execution__
Conditions can be defined with the _if_, _else_ and _elif_ keywords. In Python, the code of conditional blocks, loops, functions and classes are indented with spaces instead of using the classical C-based bracket syntax.

```python
x = 2

if x < 2:
  print("The value of x is smaller than 2.")
elif x == 2:
  print("X equals 2.")
else:
  print("X is greater than 2.")
```

There are useful keywords for logical expressions

```python
a = 5
b = 4
print(a is b)

#print(5 is not 4)  # Not good. is
print(not (5 == 4))
print((a == 5) and (b < 5))
print((False and not False) or True)
```

###__1.4. Loop execution__

Two type of loops are used in Python: _for_ and _while_ loops

In the header of _for_ loops, an iterable object has to be given. The built-in _range_ function is the easiest way to define such an object. The common arguments of range function are:
- Giving one (_n_) argument, the for loop iterates through the [0, _n_) interval stepping 1 between each period.
- Giving two (_n_, _m_) arguments, the loop iterates through the [_n_, _m_) interval stepping 1 between each period.
- Giving three (_n_, _m_, _k_) arguments, the loop iterates through the [_n_, _m_) interval stepping _k_ between each period

```python
for i in range(4):
  print(i)
print()

for i in range(2, 4):
  print(i)
print()

for i in range(2, 8, 2):
  print(i)
```

```python
a = ['b', 'c', 'd']

for element in a:
  print(element)
```

In the header of _while_ loops a logical statement takes place similarly to C which determines the condition of code execution inside the loop.

```python
i = 0
while i < 10:
  print(i)
  i += 1  # There are no -- and ++ operators in Python
```

An interesting and sometimes really convenient tool to work with lists: _list comprehension_
Some other extras:
- _sum_ is a built-in function that calculates the sum of a list.
- with _zip_ we can easily iterate through multiple data structures with the same length in one loop

```python
A = [1, 2, 3]
B = [4, 5, 6]

x = [a*b for a, b in zip(A, B)]

print(x)
print(sum(x))
```

###__1.5. Functions__

Functions and class methods are defined with the _def_ keyword.

```python
def fibonacci(n, *args, **kwargs):
  a = []  # Definition of an empty list
  a.append(0)
  a.append(1)

  for i in range(2, n):
    a.append(a[i-1]+a[i-2])

  # Extra positional parameters
  for arg in args:
    print(arg)

  # Extra named parameters
  for v in kwargs:
    print(kwargs[v])

  return a

print(fibonacci(10, 20, 30, 40, named_param1=100))
```

###__1.6. Classes__

The definition of a class is done with the _class_ keyword.

The first parameter of a class method is always a reference to the object itself, through which we can access properties and other methods of the object. Conventionally, the name of this reference _self_. However, it is not a compulsory name or keyword so other names (i.e. _this_ as used in other high-level languages) are possible too. The \_\_init\_\_ method is called at instantiation, playing the role of the constructor.

Operators can be implemented in a class. The list of the possible can be found in the [Python documentation](https://docs.python.org/3/library/operator.html). Such an operator can be the addition, object indexing or getting the length with the built-in _len_ function.

```python
# Without a parent class brackets can be omitted
class Point():
  def __init__(self, x:float, y:float):  # Types can be prescribed for arguments using colons
    self.x = x
    self.y = y

  def __add__(self, p):
    return Point(self.x+p.x, self.y+p.y)

  def add(self, p):
    self.x += p.x
    self.y += p.y

  def __str__(self):  # Used for custom printing
    return "[x:" + str(self.x) + ", y:" + str(self.y) + "]"


# Instantiation
p1 = Point(1, 2)
p2 = Point(3, 4)
p3 = p1 + p2

# Operator usage
print("p1:", p1)
print("p2:", p2)
print('--------------')
print("p3:", p3)

print("\n..............")
# Basic method usage
p1.add(p2)
print("p1:", p1)
```

###__2.1. Creating Arrays__
NumPy arrays have a restriction compared to normal Python lists that all elements of the array must be of the same type.

```python
import numpy as np

a = np.array([0, 1, 2, 3])
print(a)
print(a.shape)  # getting array dimensionality

a = np.arange(5)
b = np.arange(5)

# Expanding the array with a one long extra dimension
a = a[:, None]
b = b[None, :]
print()
print(a.shape)
print(b.shape)
```

###__2.2. Working with arrays__

```python
a = np.arange(5)
print(a)
print(2*a)

a = np.arange(5)
b = np.arange(5)
print()
print(a+b)
print(a*b)

a = np.arange(5)
b = np.arange(5)

a = a[:, None]
b = b[None, :]

# Matrix multiplication
c = a @ b
print()
print(c)
```

###__2.3. Data visualization__

```python
import matplotlib.pyplot as plt

x = np.arange(-np.pi, np.pi,  0.1)
y1 = np.sin(x)
y2 = np.sin(2*x)

plt.plot(x, y1)
plt.show()  # Not compulsory but flushes the plot cashe
```

```python
plt.plot(x, y1)
plt.plot(x, y2)  # Without plt.show() the two plots are placed on the same figure.
plt.show()
```

```python
# With the label argument and the calling of plt.legend() we can add a legend to the figure
plt.plot(x, y1, label="y1")
plt.plot(x, y2, label="y2")
plt.legend()
plt.title("Detailed figure")  # Title of the figure
plt.xlabel("Time")  # label of x axis
plt.ylabel("Signal")  # label of y axis
plt.grid(True)  # Drawing grid on the figure
plt.show()
```

###__2.4. Random number generation__

```python
from numpy import random

# Uniform distribution
a = random.rand(10)
print(a)
# Normal distribution
b = random.randn(10)
print(b)

plt.plot(a); plt.title("Uniform"); plt.show()
plt.plot(b); plt.title("Normal"); plt.show()
```

####__To visualize distributions, histograms are much more illustrative.__

```python
# Distributions are much more visible with more elements.
a = random.rand(1000)
b = random.randn(1000)

plt.hist(a); plt.title("Uniform"); plt.show()
plt.hist(b); plt.title("Normal"); plt.show()

# placing plots beside each other
plt.subplot(1, 2, 1)
plt.hist(a, np.arange(0, 1, 0.1))  # Specifying the resolution of the histogram
plt.title("Uniform")
plt.subplot(1, 2, 2)
plt.hist(b, np.arange(-3, 3, 0.1))
plt.title("Normal")
plt.suptitle("Distributions with subplot")
plt.show()
```

### Reading with OpenCV

```python
import cv2

I = cv2.imread(image_file.name)

plt.imshow(I)
plt.show()
```

The problem here is that OpenCV reads the colors in BGR format. To display the image correctly, it must be converted to RGB, so matplotlib will interpret it well.

### Color conversion

```python
I_rgb = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)

plt.imshow(I_rgb); plt.show()

I_gray = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)

# It is sometimes recommended to set the colormap for grayscale images.
plt.imshow(I_gray, cmap="gray"); plt.show()
```
